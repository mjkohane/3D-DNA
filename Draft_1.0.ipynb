# Open File

```python
File_object = open("DNA_Sequence1.txt")
txt = File_object.read()
print(len(txt))
File_object.close()
```

```python
print(txt)
print(len(txt))
```

# First Dimension
---------------------------------
 ## Base Count

```python
iA = 0
iT = 0
iC = 0
iG = 0

aCount = []
tCount = []
cCount = []
gCount = []

for base in range(len(txt)):
    if(txt[base] == "A"):
        aCount.append(iA)
        iA += 1
    elif(txt[base] == "T"):
        tCount.append(iT)
        iT += 1
    elif(txt[base] == "C"):
        cCount.append(iC)
        iC += 1
    else:
        gCount.append(iG)
        iG += 1
```

```python
print(aCount)
print(tCount)
print(cCount)
print(gCount)

print(len(aCount))
```

```python
totalCount = np.concatenate([aCount, tCount, cCount, gCount])
```

```python
print(len(totalCount))
print(totalCount)
print(sum(totalCount))
```

# Second Dimension
---------------------------------------------
## Base Index

```python
aIndex = []
tIndex = []
cIndex = []
gIndex = []

for base in range(len(txt)):
    if(txt[base] == "A"):
        aIndex.append(base)
    elif(txt[base] == "T"):
        tIndex.append(base)
    elif(txt[base] == "C"):
        cIndex.append(base)
    else:
        gIndex.append(base)
```

```python
#print(aIndex)
#print(tIndex)
#print(cIndex)
#print(gIndex)
print(len(aIndex))
print(len(tIndex))
print(len(cIndex))
print(len(gIndex))
```

```python
##Concatenate arrays
totalIndex = np.concatenate([aIndex,tIndex,cIndex,gIndex])
```

```python
print(totalIndex)
print(len(totalIndex))
```

# Third Dimension
-------------------------------------------------
## Base Proportion

```python
print(txt)
print(len(txt))
```

```python
a = []
t = []
c = []
g = []

iiA = 0
iiT = 0
iiC = 0
iiG = 0

seqLen = 1

for base in range(len(txt)):
    if(txt[base] == "A" ):
        iiA += 1
        a.append(iiA/seqLen)
        seqLen += 1
    elif(txt[base] == "T"):
        iiT += 1
        t.append(iiT/seqLen)
        seqLen += 1
    elif(txt[base] == "C"):
        iiC += 1
        c.append(iiC/seqLen)
        seqLen += 1
    else:
        iiG += 1
        g.append(iiG/seqLen)
        seqLen += 1
#seqLen += 1
```

```python
print(a)
print(t)
print(c)
print(g)
print(len(a))
print(len(t))
```

```python
totalProp = np.concatenate([a,t,c,g])
```

```python
print(totalProp)
print(len(totalProp))
#totalProp
```
